```python
text = """6..Trời xanh quen thói má hồng đánh ghen.
7..Cảo thơm lần giở trước đèn,
8..Phong tình có lục còn truyền sử xanh.
9,,Rằng năm Gia Tĩnh triều Minh,
10.. Bốn phương phẳng lặng, hai kinh vững vàng.
11..Có nhà viên ngoại họ Vương,
12..Gia tư nghĩ cũng thường thường bực trung."""

print(len(text))
```

```python
tokens = text.encode('utf-8') 
tokens = list(tokens)
tokens 
```

```python
# algorithm byte pair encoding. 
# 1. convert text to bytes. 
# 2. count frequency of each byte. 
# 3. merge most frequnent pair of bytes. 
# 4. repeat step 2 and 3 until we reach the desired vocabulary size or reach the maximum number of iterations.
# 5. convert text to tokens.
```

```python
merges = {} # (int, int) -> int 
vocab = {} # int -> byte 
vocab = {i : bytes([i]) for i in range(256)} 

num_merges = 10
new_index_bytes = 255
for _ in range(num_merges):
    # get frequency of each pair of bytes. 
    pairs = {} 
    for i, token in enumerate(tokens): 
        if i < len(tokens) - 1 : 
            pairs[(token, tokens[i+1])] = pairs.get((token, tokens[i+1]), 0) + 1
    most_frequent_pair = max(pairs, key=pairs.get)
    
    # merge most frequent pair of bytes into tokens.
    new_token = new_index_bytes + 1
    new_tokens = []
    i = 0
    while i < len(tokens): 
        if i < len(tokens) - 1 and (tokens[i], tokens[i+1]) == most_frequent_pair: 
            new_tokens.append(new_token)
            i += 2
        else: 
            new_tokens.append(tokens[i])
            i += 1
    tokens = new_tokens
    merges[most_frequent_pair] = new_token 
    print(new_token, most_frequent_pair)
    vocab[new_token] = vocab[most_frequent_pair[0]] + vocab[most_frequent_pair[1]] 
    print(merges)
    new_index_bytes = new_token

print(len(tokens))
```

```python
# decode tokens to text based on merges. 
text = """6..Trời xanh quen thói má hồng đánh ghen.
7..Cảo thơm lần giở trước đèn,
8..Phong tình có lục còn truyền sử xanh.
9,,Rằng năm Gia Tĩnh triều Minh,
10.. Bốn phương phẳng lặng, hai kinh vững vàng.
11..Có nhà viên ngoại họ Vương,
12..Gia tư nghĩ cũng thường thường bực trung."""
tokens = list(text.encode('utf-8'))
print(len(tokens)) 
for merge in merges: 
    new_tokens = []
    i = 0
    while i < len(tokens): 
        if i < len(tokens) - 1 and (tokens[i], tokens[i+1]) == merge: 
            new_tokens.append(merges[merge])
            i += 2
        else: 
            new_tokens.append(tokens[i])
            i += 1
    tokens = new_tokens
print(len(tokens))
```

```python
# decode tokens to text based on vocab.
text = b"".join([vocab[token] for token in tokens])
text.decode('utf-8')
```
